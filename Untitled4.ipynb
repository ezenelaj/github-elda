```python
import requests
from bs4 import BeautifulSoup
```

```python
r=requests.get("https://www.lotioncrafter.com/lotioncrafter-premium-ingredients-botanical-extracts/")
```

```python
c=r.content
```

```python
soup=BeautifulSoup(c,"html.parser")
```

```python
s1=soup.find_all("td",{"class":"PListCell"})
```

```python
lista=[]
for i in s1:
        d={}
        d["product"]=i.find("a",{"class":"ProductTitle"}).text
        d["price"]=i.find("span",{"style":"WHITE-SPACE: nowrap"}).text.replace("\n","").replace("$","").replace(" ","")
        lista.append(d)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(lista)
```

```python
df['price $']=pd.to_numeric(df['price'])
```

```python
df.pop('price')
```

```python
df.sort_values(by=['price $'])
```

```python
df.loc[(df['price $']>2)&(df['price $']<20)].sort_values(by=['price $'],ascending=False).head(5)
```
