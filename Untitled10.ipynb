```python
import folium
for i in range(data['Latitude'].shape[0]):
    map=folium.Map(location=[data['Latitude'].iloc[i],data['Longitude'].iloc[i]],tiles="Mapbox Bright")
fg=folium.FeatureGroup(name="My Map")
for i in range(data['Latitude'].shape[0]):
    fg.add_child(folium.Marker(location=[data['Latitude'].iloc[i],data['Longitude'].iloc[i]],popup=data['Volcano'].iloc[i],icon=folium.Icon(color='purple')))

map.add_child(fg)
map.save("Map3.html")
```
